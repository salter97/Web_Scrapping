```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://apexlegendsstatus.com/algs/Y4-Split2/ALGS-Playoffs/Global/Overview#tab-statsPlayers'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table',id='playersStatsTable')[0]
```

```python
table = soup.find_all('table',id='playersStatsTable')[0]
```

```python
print(table)
```

```python
table.find_all('th')
```

```python
apex_titles = table.find_all('th')
```

```python
world_apex_titles = [title.text.strip() for title in apex_titles ]
```

```python
world_apex_titles
```

```python
import pandas as pd
```

```python
df  = pd.DataFrame(columns= world_apex_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data ]
    print(individual_row_data)

    length = len(df)
    df.loc[length] = individual_row_data
```
